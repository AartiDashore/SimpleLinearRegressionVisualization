```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Generate synthetic data
np.random.seed(0)  # For reproducibility
X = 2 * np.random.rand(100, 1)  # 100 data points with one feature
y = 4 + 3 * X + np.random.randn(100, 1)  # Linear relation with some noise

# Create and train the linear regression model
model = LinearRegression()
model.fit(X, y)

# Predict using the model
X_new = np.linspace(0, 2, 100).reshape(100, 1)  # Generate new data points for the line
y_predict = model.predict(X_new)

# Calculate metrics
mse = mean_squared_error(y, model.predict(X))
r2 = r2_score(y, model.predict(X))

# Plotting
plt.figure(figsize=(10, 6))

# Scatter plot of the original data
plt.scatter(X, y, color='blue', label='Data points')

# Plot the regression line
plt.plot(X_new, y_predict, color='red', linewidth=2, label='Regression line')

# Plot residuals
for i in range(len(X)):
    plt.plot([X[i], X[i]], [y[i], model.predict(X[i].reshape(1, -1))[0]], color='green', linestyle='--')

plt.title('Linear Regression Visualization')
plt.xlabel('Feature')
plt.ylabel('Target')
plt.legend()
plt.grid(True)

# Show plot
plt.show()

# Print metrics
print(f"Mean Squared Error: {mse:.2f}")
print(f"R^2 Score: {r2:.2f}")
```
